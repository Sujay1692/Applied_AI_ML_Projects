```python
df= pd.read_csv('/content/drive/MyDrive/World University Rankings.csv')
df
```

```python
mapping = {'master_account': 0, 'private': 1, 'public': 2}
df['record_type'] = df['record_type'].map(mapping)
```

```python
df.dropna()
```

```python
df=df.dropna()
df
```

```python
x = df[['scores_teaching','scores_research','scores_citations','scores_industry_income']]
y = df[['record_type']]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,stratify=y,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=15) # creat model
knn.fit(x_train, y_train) # fitting the model based on the train samples
y_predict = knn.predict(x_test) # predict model based on test sample
y_predict
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=6, random_state=42) # Creat Model
dtc.fit(x_train, y_train) # Fit the model
```

```python
predict_dtc = dtc.predict(x_train) # Prediction
```

```python
predict_dtc
```

```python
dtc.predict(x_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_train, predict_dtc)
```

```python
dtc.score(x_test,y_test)
```

```python
xnew=[1,1,1,1]
predict2= dtc.predict([xnew])
predict2
```

```python
# visualisation of decision tree

import graphviz
from sklearn.tree import export_graphviz
export_graphviz(dtc, out_file="tree.dot",filled=True, rounded=True)
```

```python
# Load and render the decision tree from the dot file
with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```
