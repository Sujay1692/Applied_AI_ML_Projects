```python
import statsmodels.formula.api as smf
```

```python
riskmodel = smf.logit(formula = 'gender ~ height_cm + weight_kg + diastolic + systolic + gripForce + sit_and_bend_forward_cm + broad_jump_cm', data = df) # create model

binary=riskmodel.fit() # fit model
```

```python
binary.summary()
```

```python
coef=binary.params
print(np.exp(coef))
```
