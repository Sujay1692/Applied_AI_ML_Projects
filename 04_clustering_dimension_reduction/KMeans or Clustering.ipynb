```python
df= pd.read_csv('/content/drive/MyDrive/World University Rankings.csv')
df
```

```python
df.dropna()
```

```python
df=df.dropna()
```

```python
kmn = KMeans(n_clusters=6)
kmn.fit(df[['scores_teaching','scores_research','scores_citations','scores_industry_income','scores_international_outlook']]) # fitting data - the input of fitting is a DataFrame with its columns
Clusters = kmn.predict(df[['scores_teaching','scores_research','scores_citations','scores_industry_income','scores_international_outlook']]) # prediction : show us the cluster which maschine found
```

```python
df['cluster1']=Clusters
df
```

```python
xs = df['scores_teaching']
ys = df['scores_research']

labels = Clusters
centroids = kmn.cluster_centers_ # this is very important function to see the place of center from clustering
plt.scatter(xs, ys, c=labels)
plt.xlabel('scores_teaching')
plt.ylabel('scores_research')
plt.scatter(centroids[:,0],centroids[:,2],marker='x', linewidth=10, s=450,c='red')
plt.show()
```
