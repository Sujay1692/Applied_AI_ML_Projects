# Line PLot

```python
plt.plot(df['scores_teaching'],df['scores_research'])
plt.title("Teaching vs Research Scores")
plt.xlabel("scores_teaching")
plt.ylabel("scores_research")
plt.show()
plt.show()
```

## Scatter PLot

```python
plt.scatter(df['scores_teaching'],df['scores_research'])
plt.title("Teaching vs Research Scores")
plt.xlabel("Teaching scores")
plt.ylabel("Research scores")
plt.show()
```

## Bar Plot

```python
mean_teaching=df['scores_teaching'].mean()
mean_research=df['scores_research'].mean()

t1=[1]
t2=[2]

tt1=[mean_teaching]
tt2=[mean_research]

plt.bar(t1, tt1, label="Teaching", color='red')
plt.bar(t2, tt2, label="Research", color='green')

plt.title("Mean Teaching vs Research Scores")
plt.xlabel("Category")
plt.ylabel("Mean Score")
plt.show()
```

# Pie Plot

```python
mean_teaching = df['scores_teaching'].mean()
mean_research = df['scores_research'].mean()
mean_citations = df['scores_citations'].mean()
mean_industry = df['scores_industry_income'].mean()

labels = ['Teaching', 'Research', 'Citations', 'Industry Income']
sizes = [mean_teaching, mean_research, mean_citations, mean_industry]
colors = ['red', 'green', 'blue', 'yellow']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Mean Scores Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,6), dpi=100)
plt.hist(df['scores_teaching'], color='skyblue', label='Teaching Scores')
plt.hist(df['scores_research'], color='green', label='Research Scores')
#plt.hist(df['scores_citations'], color='red', label='Citations Scores')
#plt.hist(df['scores_industry_income'], color='yellow', label='Industry Income Scores')
plt.xlabel('Score')
plt.ylabel('Number of University')
plt.title('Distribution of Scores')
plt.legend()
plt.show()
```
